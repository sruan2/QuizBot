```python
import numpy as np
import matplotlib
#import matplotlib.pyplot as plt
from matplotlib import pyplot
%matplotlib inline
```

```python
# prettier plot
plt.style.use('ggplot')
```

```python
quizbot=[59.8, 78.62, 49.23, 45.81]
flashcard = [33.86, 41.84]
bins = np.linspace(30, 80, 5)

pyplot.hist(quizbot, bins, alpha=0.5, label='quizbot')
pyplot.hist(flashcard, bins, alpha=0.5, label='flashcard')
pyplot.legend(loc='upper left')
pyplot.title("Plot")
pyplot.show()
```

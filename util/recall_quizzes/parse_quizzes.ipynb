```python
# read in the csv file 
df_pre = pd.read_csv('Quiz A MTurk1 10 - 5 Pre_August 22, 2018_16.47.csv', index_col='Qualtrics ID')
df_post = pd.read_csv('Quiz B MTurk1 10_5 Post_August 26, 2018_15.22.csv', index_col='Qualtrics ID')
df_pre.insert(0, 'QID', new_pre_qualtrics_to_qid)
df_post.insert(0, 'QID', new_post_qualtrics_to_qid)
df_pre[:2]
# df_post[:2]
```

```python
# generate the quizzes
for name in df_pre.columns[2:-1]:
    print(name)
    name_df = df_pre[['QID', 'Question', name, 'Answer']]
    incorrect_answers = name_df[name_df[name] != name_df['Answer']]
    thing = incorrect_answers[['QID', 'Question', 'Answer']]
    thing.to_csv(name + '_custom_quiz.csv')
```

```python
# list of existing ids in the database, use to find missing ones to append in order
qs= [1,2,3,5,6,8,9,10,11,14,15,17,18,19,21,24,25,27,29,30,31,32,35,37,38,40,41,42,44,45,46,49,50,51,52,53,55,56,58,59,64,65,68,70,72,73,74,75,76,77,79,82,83,84,85,87,88,89,90,92,93,94,97,99,100,102,103,104,106,111,112,113,114,117,118,120,121,125,127,128,130,131,132,133,135,136,138,139,141,142,143,144,145,146,147,148]
missing_qs_pre = pd.Series(list(set(qs) - set(df_pre['QID'])), name = 'QID')
missing_qs_post = pd.Series(list(set(qs) - set(df_post['QID'])), name = 'QID')
print(len(missing_qs_pre))
print(len(missing_qs_post))
df_pre = df_pre.set_index('QID')
df_post = df_post.set_index('QID')
```

```python
# append columns to the spreadsheet
pre_cols = df_pre.columns
extended_df_pre = df_pre.append(pd.DataFrame('', index=missing_qs_pre, columns=pre_cols))
extended_df_pre = extended_df_pre.sort_index()

post_cols = df_post.columns
extended_df_post = df_post.append(pd.DataFrame('', index=missing_qs_post, columns=post_cols))
extended_df_post = extended_df_post.sort_index()

final_df = extended_df_pre[['Question']]
final_df.loc[df_post.index, 'Question'] = df_post['Question']

for name in extended_df_pre.columns[1:-1]:
    #final_df.loc[:,name+' pre'] = ''
    #final_df.loc[:,name+' post'] = ''
    final_df.loc[df_pre.index,name+' pre'] = (df_pre['Answer'] == df_pre[name]).astype(int)
for name in extended_df_post.columns[1:-1]:
    final_df.loc[df_post.index,name+' post'] = (df_post['Answer'] == df_post[name]).astype(int) \
                                                if name in df_post.columns else ''
```

```python
final_df= final_df.replace(np.nan, '', regex=True)
```
